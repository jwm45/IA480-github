```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import psycopg2

conn = psycopg2.connect(host=get_secret('postgresql')['host'],
                      user=get_secret('postgresql')['username'],
                      password = get_secret('postgresql')['password'],
                      dbname=get_secret('postgresql')['engine'])

cur = conn.cursor()
```

```python
sql = " select * from name inner join income on name.fips = income.fips "
```

```python
import pandas as pd

df = pd.read_sql(sql,conn)
df.head()
```

```python
%%ai gpt4
what is in the {df} in {In[10]}
```

```python
%%ai gpt4 -f code

calculate the income growth rate for all states from the {df} in {In[10]}
display the result on chat
```

```python
# first, remove the duplicate 'fips' column
df = df.loc[:,~df.columns.duplicated()]

# pivot the dataframe to have years as columns and calculate the growth rate
df_pivot = df.pivot_table(values='income', index='name', columns='year')
df_growth = df_pivot.pct_change(axis='columns')

# display the growth rate from 2005 to 2019 for each state
growth_rate = df_growth.loc[:, [2005, 2019]]
print(growth_rate)
```

```python
%%ai gpt4 -f code
Calculate the average year-over-year income growth rate for each state in {df} 
and create a chart to show the result
```

```python
import matplotlib.pyplot as plt

# pivot the dataframe to have years as columns
df_pivot = df.pivot_table(values='income', index='name', columns='year')

# calculate the year-over-year growth rate
df_growth = df_pivot.pct_change(axis='columns')

# calculate the average growth rate
average_growth_rate = df_growth.mean(axis=1)

# create a bar chart to show the average growth rate for each state
average_growth_rate.sort_values().plot(kind='bar', figsize=(15, 7))
plt.ylabel('Average Year-Over-Year Income Growth Rate')
plt.show()
```
